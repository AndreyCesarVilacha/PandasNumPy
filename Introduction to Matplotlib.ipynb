## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting injupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show
fig.savefig("images/sample-plot.png")
```
